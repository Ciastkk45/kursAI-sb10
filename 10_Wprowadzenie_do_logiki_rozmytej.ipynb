```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 50, 500)

def fuzzy_comfort(x):
  return np.maximum(np.minimum(np.minimum((x-10)/10, 1), (40 - x) / 10), 0)
y = fuzzy_comfort(x)

points_x = [10, 20, 30, 40]
points_y = fuzzy_comfort(np.array(points_x))
labels = ["a (10st)", "b (20st)", "c (30st)", "d (40st)"]

plt.plot(x, y, label="Temperatura komfortowa")
plt.scatter(points_x, points_y, color="black")
for i, label in enumerate(labels):
  plt.annotate(label, (points_x[i], points_y[i]+0.1))
plt.legend()
plt.grid()
plt.xlabel("Temperatura (st C)")
plt.ylabel("Przynależność")
plt.title("Funkcja przynależności: Temperatura komfortowa (trapezowa: 10, 20, 30, 40)")
plt.show
```
