```python
#Wyznaczenie ekstremów (minimów i maksimów) i wartości średnich dla znanego zbioru treningowego

# Import bibliotek
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# Załaduj dane Iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Podział na 70 % treningowych i 30 % testowych
df_train, df_test = train_test_split(df, test_size=0.30, random_state=42)

# Lista cech
features = [
    "sepal length (cm)",
    "sepal width (cm)",
    "petal length (cm)",
    "petal width (cm)"
]

# Gatunki
species = iris.target_names

# Tworzenie tabeli wyników
results = []
for feature in features:
    for specie in species:
        data = df_test[df_test['species'] == specie][feature]
        results.append({
            "Feature": feature,
            "Species": specie,
            "Minimum": data.min(),
            "Maximum": data.max(),
            "Mean": data.mean()
        })

# DataFrame z wynikami
summary_df = pd.DataFrame(results)

# Wyświetlenie tabeli
print(summary_df)
```

```python
import numpy as np
import matplotlib.pyplot as plt

feature = "petal length (cm)"
specie = "setosa"

row = summary_df[(summary_df["Feature"] == feature) & (summary_df["Species"] == specie)].iloc[0]
min_val = row["Minimum"]
mean_val = row["Mean"]
max_val = row["Maximum"]

def funkcja_przynaleznosci(x, a, b, c):
  if x <= a or x >= c:
        return 0
  elif a < x < b:
        return (x - a) / (b - a)
  elif b <= x < c:
        return (c - x) / (c - b)

x_values = np.linspace(min_val - 0.5, max_val + 0.5, 200)
y_values = [funkcja_przynaleznosci(x, min_val, mean_val, max_val) for x in x_values]

plt.plot(x_values, y_values, label=f"Funkcja przynależności\n{specie}, {feature}")
plt.xlabel(feature)
plt.ylabel("Przynależność")
plt.title("Przykładowa funkcja przynależności (trójkątna)")
plt.grid(True)
plt.legend()
plt.show()
```
